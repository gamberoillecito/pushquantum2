```python
from scipy.optimize import minimize as min
import numpy as np
import numpy.random as r
from numpy import asarray
from matplotlib import pyplot as plt
```

$\phi = f(\vec{\theta})$,
where
$\vec{\theta} = S\{argmin_\theta C_f(\vec{\theta})$\}

```python
def C_prova(theta):
    theta = asarray(theta)
    sigA = r.rand(2,2)
    sigB = r.rand(2,2)
    rho = r.rand(2,2)
    return np.sum((np.trace(np.sqrt(sigA)*rho**np.sqrt(sigB)))**2 + 1/2*(1+theta[1:])/(1-theta[:-1]))
```

```python
theta0 = [] #output del nostro quantum computer, in termini di parametri
th_num = 12 #number of theta parameters
theta0_prova = r.rand(th_num) #
num = 100 #number of iterations

res = []
for i in range(num):
    sol = min(C_prova, theta0_prova, method='Nelder-Mead', tol=1e-6)
    res.append(sol.x)
```

```python
plt.cla()
plt.clf()
plt.close()
R = []
for j in range(len(res[1])):
    S = []
    for i in range(num):
        S.append(res[i][j])
    R.append(S)
    plt.plot(range(num),R[j])
```
